```python
# Import the pathlib and csv library
from pathlib import Path
import csv
```

```python
# Set the file path
csvpath = Path("Homework_Week 2_PyBank_Resources_budget_data.csv")
```

```python
# Initialize dictionary
analysis = {}
```

```python
# Open the csv file as an object
with open(csvpath, "r") as csvfile:

    # Pass in the csv file to the csv.reader() function
    # (with ',' as the delmiter/separator) and return the csvreader object
    csvreader = csv.reader(csvfile, delimiter=",")

    # Read the header row first (skip this step if there is no header)
    csv_header = next(csvreader)
    # Print the header
    print(csv_header)
    
    #initiate counters and lists
    total_months = 0
    total_profit = 0
    previous_months_profit = 0
    monthly_change_lst = []
    greatest_increase = 0
    greatest_increase_month = ""
    greatest_decrease = 0
    greatest_decrease_month = ""
    
    # Read each row of data after the header
    for row in csvreader:
        # Print the row
        print(row)
        
        # Set the 'date' and 'monthly_return' variables for better
        # readability, convert strings to ints for numerical calculations
        month = row[0]
        monthly_return = int(row[1])
        
        total_months += 1
        total_profit += monthly_return
        
        if total_months == 1:
            previous_months_profit = monthly_return
            print ()
        elif total_months > 1:
            monthly_change = monthly_return - previous_months_profit
            print (f"change from last month: {monthly_change}")
            monthly_change_lst.append(monthly_change)
            previous_months_profit = monthly_return
            print ()
        
        if greatest_increase == 0:
            greatest_increase = monthly_change
        elif monthly_change > greatest_increase:
            greatest_increase = monthly_change
            greatest_increase_month = month
        
        if greatest_decrease == 0:
            greatest_decrease = monthly_change
        elif monthly_change < greatest_decrease:
            greatest_decrease = monthly_change
            greatest_decrease_month = month
        
```

```python
# The total number of months included in the dataset.
print (f"Total months: {total_months}")
```

```python
 # The net total amount of Profit/Losses over the entire period.
print (f"Total profit over the period: ${total_profit}")
```

```python
 #The average of the changes in Profit/Losses over the entire period.
print ("prove that monthly changes are being added to a list:")
print (monthly_change_lst)
print ("_______________")
print ()
average_monthly_change = round(sum(monthly_change_lst) / len(monthly_change_lst),2)
print (f"Average Monthly Change: ${average_monthly_change}")
```

```python
#The greatest increase in profits (date and amount) over the entire period.
print (f"Greatest Increase in Profits: {greatest_increase_month} {greatest_increase}")        
```

```python
#The greatest decrease in profits (date and amount) over the entire period.
print (f"Greatest Decrease in Profits: {greatest_decrease_month} {greatest_decrease}")   
```

```python
# Your final script should print the analysis to the terminal and export a text file with the results.
print_rows = []
row1 = "Financial Analysis"
row2 = ("----------------------------")
row3 = (f"Total months: {total_months}")
row4 = (f"Total profit over the period: ${total_profit}")
row5 = (f"Average Monthly Change: ${average_monthly_change}")
row6 = (f"Greatest Increase in Profits: {greatest_increase_month} {greatest_increase}")   
row7 = (f"Greatest Decrease in Profits: {greatest_decrease_month} {greatest_decrease}")   
print_rows.append(row1)
print_rows.append(row2)
print_rows.append(row3)
print_rows.append(row4)
print_rows.append(row5)
print_rows.append(row6)
print_rows.append(row7)
print (print_rows)
```
